```python
import pandas as pd

pd.options.display.max_columns = 100
pd.options.display.float_format = '{:,.2f}'.format
```

```python
df = pd.read_csv("data/2-datasets/H2020_organizations.csv", na_values="-")
```

```python
df.groupby(["country"])["netEcContribution"].agg(['sum','count', 'median']).nlargest(10, 'sum')
```

```python
pd.pivot_table(df, index="name", columns="role", values="netEcContribution", aggfunc="sum" ).nlargest(10, "thirdParty")
```
